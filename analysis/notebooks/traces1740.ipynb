```python
import ROOT
import numpy as np
import matplotlib.pyplot as plt
```

```python
inFile = ROOT.TFile.Open ( "../huntingTraces/zooTraces100binsPMT1.root" ," READ ")
```

```python
hist = inFile.Get ("badTrac")
histok = inFile.Get("pmthok")
```

```python
arr0 = []
arr1 = []
for e in range(0, hist.GetNbinsX()):
    if hist.GetBinContent(e, 1) > 0:
        arr0 = []
        for i in range( 0, hist.GetNbinsY()+1):
            arr0.append( hist.GetBinContent(e, i) )
        arr1.append( arr0 )
```

```python
for i in range(0, len(arr1)):
    plt.plot(arr1[i])
    
plt.xlim(1,)
plt.xlabel("Bins (from 0 to 99, first bins; from 100 to 200, last bins)")

plt.ylim(100,280)
plt.ylabel("FADC")

plt.title( "Not Ok Traces for Station 1740\n("+str(len(arr1))+"/608)" )
plt.grid()
plt.savefig("../../plots/traces1740.pdf", dpi=300)
```
